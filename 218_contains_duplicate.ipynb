First solution with sorting and then looping over all the elements. Time complexity O(nlogn)

```python
class Solution(object):
    def containsDuplicate(self, nums):
        """
        :type nums: List[int]
        :rtype: bool
        """
        nums.sort()
        if len(nums) == 1:
            return False
        else:
            for i in range(0, len(nums)-1):
                if nums[i] == nums[i+1]:
                    return True
            if nums[-2] == nums[-1]:
                return True
            else:
                return False
```

Second solution using hashsets, time complexity O(n)

```python
class Solution(object):
    def containsDuplicate(self, nums):
        """
        :type nums: List[int]
        :rtype: bool
        """
        hashset = set()

        for i in nums:
            if i in hashset:
                return True
            else:
                hashset.add(i)
        return False
```
